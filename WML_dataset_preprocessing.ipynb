```python
import os
import cv2
```

```python
DirpathObj="ExDark_Annno\\People"
DirpathObj1="training\\dataset\\Hack_IBM\\labels\\train"
DirpathImg="ExDark\\People"
```

```python
for filename in os.listdir(DirpathObj):
    path=os.path.join(DirpathObj,filename)
    path1=os.path.join(DirpathObj1,filename[:-8]+".txt")
    pathImg=os.path.join(DirpathImg,filename[:-4])
    img=cv2.imread(pathImg)
    height=img.shape[0]
    width=img.shape[1]
    dw=1/width
    dh=1/height
    print(filename)
    final=[]
    with open(path, 'r') as G:
        data=G.readlines()
    data.remove(data[0])
    for i in data:
        x=i.split(" ")
        if x[0]=='Bicycle':
            x[0]='0'
        if x[0]=='Bus':
            x[0]='1'
        if x[0]=='Car':
            x[0]='2'
        if x[0]=='Cat':
            x[0]='3'
        if x[0]=='Dog':
            x[0]='4'
        if x[0]=='Motorbike':
            x[0]='5'
        if x[0]=='People':
            x[0]='6'
        W=int(x[3])
        H=int(x[4])
        X=(int(x[1])+W/2.0)
        Y=(int(x[2])+H/2.0)
        X=X*dw
        W=W*dw
        Y=Y*dh
        H=H*dh
        data1=x[0]+" "+str(X)+" "+str(Y)+" "+str(W)+" "+str(H)+"\n"
        final.append(data1)
    print(final)
    with open(path1, 'w') as G1:
        data=G1.writelines(final)
```
